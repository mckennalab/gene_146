### Working with Dictionaries

Dictionaries are useful for storing and retrieving data as key-value pairs.  For example, here is a short dictionary of masses. The keys are dNTPs, and the values are the corresponding masses.

```python
mw = {'dATP': 491.2, 'dCTP': 467.2, 'dGTP':507.2, 'dTTP': 482.2}
mw
```

We can a value to an existing dictionary.

```python
mw['dMagic'] = 114.23
mw
```

We can retrieve a value from a dictionary.

```python
mw['dATP']
```

A for loop is a useful means of interating over all key-value pairs of a dictionary.

```python
for species in mw.keys():
    print("The molar mass of {:<s} is {:<7.2f}".format(species, mw[species])) # more fancy printing options
```

Dictionaries can be sorted by key or by value

```python
for species in sorted(mw):
    print(" {:<8s}  {:>7.2f}".format(species, mw[species]))
```

```python
for species in sorted(mw, key = mw.get):
    print(" {:<8s}  {:>7.2f}".format(species, mw[species]))
```

## Functions

A function defines a bit of code that takes input parameters and optionally returns a value or values

```python
def smaller_number(first_number, second_number):
    if first_number < second_number:
        return(first_number)
    else:
        return(second_number)
    
smaller_number(10,5)
```

### Plotting with Matplotlib

Importing the `matplotlib.pyplot` library gives IPython notebooks plotting functionality very similar to Matlab's. Here are some examples using functions from the 

```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0,10)
y = np.sin(x)
z = np.cos(x)

plt.plot(x,y,'b',x,z,'r')
plt.xlabel('Radians');
plt.ylabel('Value');
plt.title('Plotting Demonstration')
plt.legend(['Sin','Cos'])
plt.grid()
```

### Nicer, more 'ggplot level' figures
